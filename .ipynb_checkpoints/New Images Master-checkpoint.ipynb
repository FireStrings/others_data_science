```python
# importing libraries 
import matplotlib.pyplot as plt 
import seaborn 

# declaring data 
data = [49, 25, 15, 6, 5] 
labels = ['Metereological records', 'NWP', 'PV power history data', 
          'Satellite cloud images', 'Sky images'] 

# declaring exploding pie 
explode = [0.05, 0.05, 0.05, 0.05, 0.05] 
# define Seaborn color palette to use 
palette_color = seaborn.color_palette('Accent') 

# plotting data on chart 
ax, ax2, ax3 = plt.pie(data, labels=labels, colors=palette_color, 
explode=explode, autopct='%.0f%%') 

# plt.legend(bbox_to_anchor=(0.3,0.5), loc="center right", 
#                           bbox_transform=plt.gcf().transFigure)

plt.show() 

plt.savefig("out1.png")
```
