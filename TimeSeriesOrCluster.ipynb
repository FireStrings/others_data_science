```python
import pandas as pd
import numpy as no
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("data/weatherAUS.csv")
```

```python
data.head()
```

```python
data['data'] = pd.to_datetime(data['Date'])
data2 = data.groupby("data").agg("mean")
```

```python
ax = sns.lineplot(x='data', y='MaxTemp', data=data2.iloc[0:1000])
sns.set_palette("Accent")
sns.set_style("darkgrid")
ax.figure.set_size_inches(12,6)
ax.set_title("Temperaturas", loc='left', fontsize=18)
ax.set_xlabel("Tempo", fontsize=14)
ax.set_ylabel("Temp", fontsize=14)
ax = ax
```

```python
data2['aumento'] = data2.MaxTemp.diff()
```

```python
ax = sns.lineplot(x='data', y='aumento', data=data2)
ax.figure.set_size_inches(12,6)
ax.set_title("Aumento das Temperaturas", loc='left', fontsize=18)
ax.set_xlabel("Tempo", fontsize=14)
ax.set_ylabel("Aumento", fontsize=14)
ax = ax
```
